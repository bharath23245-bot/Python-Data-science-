```python
#PLACEMENT SYSTEM VISUALISATION
import matplotlib.pyplot as plt
import numpy as np
years=np.array([2016,2017,2018])
cse_list=[]
it_list=[]
ece_list=[]
mech_list=[]
for i in range(3
               ):
  cse=int(input("ENTER THE CSE STUDENTS:"))
  it=int(input("ENTER THE IT STUDENTS:"))
  ece=int(input("ENTER THE ECE STUDENTS:"))
  mech=int(input("ENTER THE MECHANICAL STUDENTS:"))
  cse_list.append(cse)
  it_list.append(it)
  ece_list.append(ece)
  mech_list.append(mech)
plt.bar(years,cse_list,label="cse")
plt.bar(years,it_list,label="it",bottom=cse_list)
plt.bar(years,ece_list,label="ece",bottom=it_list)
plt.bar(years,mech_list,label="mech",bottom=ece_list)
plt.xlabel("Years")
plt.ylabel("Number of Students Placed")
plt.title("Placement System Visualization")
plt.legend()
plt.show()
```
